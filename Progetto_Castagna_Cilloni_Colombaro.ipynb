```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
dogs=dogs[dogs['description'].notna()]
```

```python
duplicate=dogs[['id','breed_primary', 'breed_secondary','sex']].groupby(['breed_primary','breed_secondary','sex']).count()
same_breed=pd.DataFrame()
same_breed=duplicate[duplicate['id']>1]
same_breed=same_breed.reset_index()
same_breed=same_breed.rename(columns={'id':"count"})
dict_elem={}
for i in range(len(same_breed)):
    dict_elem[i]=dogs[(dogs['breed_primary']==same_breed.iloc[i]['breed_primary']) & (dogs['breed_secondary']==same_breed.iloc[i]['breed_secondary']) & (dogs['sex']==same_breed.iloc[i]['sex'])] ['id']
```

```python
for i in range(len(dict_elem)):
    dict_elem[i]=dict_elem[i].values.tolist()
same_breed['caniId']=dict_elem.values()
```

```python
def similarity(X,Y):
    #list for the values in the string
    X_list = word_tokenize(X)  
    Y_list = word_tokenize(Y)
    l1 =[]
    l2 =[]
    union=X_list+Y_list
    #l1 : 1= if the element is in the list, 2= if the element is not in the list
    #l2:  1= if the element is in the list, 2= if the element is not in the list
    for w in union:
        if w in X_list:
            l1.append(1)
        else:
            l1.append(0)
        if w in Y_list: 
            l2.append(1)
        else: 
            l2.append(0)
    # cos formula 
    c = 0
    for i in range(len(union)):
            c+= l1[i]*l2[i]
    #cosine = c / float((sum(l1)*sum(l2))**0.5)
    d1=0
    for i in l1:
        d1+=i**2
    d1=d1**0.5
    d2=0
    for i in l2:
        d2+=i**2
    d2=d2**0.5
    cos = c / float(d1*d2)
    return(cos)
```

```python
values=[] 
for i in same_breed['caniId']:
        for j in range(0,len(i)-1):
            for t in range(len(i)-1,j,-1):
                x=str(dogs['description'][dogs['id']==i[j]])
                y=str(dogs['description'][dogs['id']==i[t]])
                if similarity(X=x,Y=y)>=0.9:
                    values.append((i[j],i[t]))
print(f'The number of pairs of dogs that are equal to each other is {len(values)}')
```
